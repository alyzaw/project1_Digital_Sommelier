```python
# Dependencies
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt


from pprint import pprint

# Google developer API key
import sys
sys.path.append('../..')
from config import gkey


file = pd.read_csv("winedata_coord_type_2019JAN21.csv")

file.head(30)
```

```python
import gmaps
import numpy as np


# Configure gmaps with API key
gmaps.configure(api_key=gkey)


# Store 'Lat' and 'Lng' into  locations 
locations = file[["Latitude", "Longitude"]].astype(float)
unique_locations = locations.drop_duplicates()
```

```python
# Create winery symbol layer
winery_layer = gmaps.marker_layer(unique_locations)


fig = gmaps.figure()
fig.add_layer(winery_layer)

fig

```
