```python
# Dependencies

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load Data

df = pd.read_csv("winedata_coord_type.csv")
df.head(2)
```

```python
#Calculate how many wines are in the dataset
CountofWines = df["Title"].count()

# Print out the number of wines in the dataset. 
print("There are " + str(CountofWines) + " in this dataset.")
```

```python
CountofCountries = df["Country"].value_counts()
CountofCountries
```

```python
#Create dataframes to analyze numbers

grade =df["Grade"]
price = df["Price"]
```

```python
#Finding Prices of Wine of the world

max_price=price.max()
min_price=price.min()
median_price=price.median()
avg = price.sum()/CountofWines


# Converting string to integer
#int_max_price = int(max_price)
#int_min_price = int(min_price)

print("The most expensive wine cost $" + str(max_price) + " in this dataset.")
print("The cheapest wine cost $" + str(min_price) + " in this dataset.")
print("The median cost of wine is $" + str(median_price) + " in this dataset.")
print("The average cost of wine is $" + str(avg) + " in this dataset.")
print("===================================================================")


# Additional price numbers
ten_pctille = (np.percentile(price,10))
lower_quartile = (np.percentile(price,25))
median = (np.percentile(price,50))
upper_quartile = (np.percentile(price,75))
ninety_pctille = (np.percentile(price,90))

print("Cost of wine at the 10th percentile $" + str(ten_pctille))
print("Cost of wine at the 25th percentile $" + str(lower_quartile))
print("Cost of wine at the 50th percentile $" + str(median))
print("Cost of wine at the 75th percentile $" + str(upper_quartile))
print("Cost of wine at the 90th percentile $" + str(ninety_pctille))

```

```python
#Finding Grades of Wine of the world

max_grade=grade.max()
min_grade=grade.min()
median_grade=grade.median()

print("The maximum grade is: " + str(max_grade))
print("The minimum grade is: " + str(min_grade))
print("The median grade is: " + str(median_grade))
```

```python
# Create a group based on the values in the 'Country' column
year_group = df.groupby('Year')

# Count how many times each country appears in our group
count_year = year_group['Year'].count()

count_year
```

```python
# Create a bar chart based off of the group series from before
count_chartyears = count_year.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chartyears.set_xlabel("Years")
count_chartyears.set_ylabel("Number of Wines")

# Save our graph and show the grap
plt.tight_layout()
plt.savefig("WineYears.png")
plt.show()
```
