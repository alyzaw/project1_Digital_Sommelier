```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# Read in data
file = pd.read_csv("winedata_coord_type.csv")


grade = file["Grade"]
year = file["Year"]
price = file["Price"]
```

```python
# Run the t-test
(t_stat, p) = stats.ttest_ind(year, price, equal_var=False)
```
